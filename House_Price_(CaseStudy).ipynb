### Q.1)find the owner and dealer whose target price below 60 lac but they should approved by RERA. 

```python
data=dataset.loc[((dataset["POSTED_BY"]=="Owner") | (dataset["POSTED_BY"]=="Dealer")) & (dataset["TARGET(PRICE_IN_LACS)"]<60) & (dataset["RERA"]==1)]
data
```

### Q.2) Find the 2 BHK owner those target price is between 60-90 lacs and that should be ready to move.

```python
data2=dataset.loc[(dataset["POSTED_BY"]=="Owner")&(dataset["BHK_NO."]==2) & ((dataset["TARGET(PRICE_IN_LACS)"]<=90.0) & (dataset["TARGET(PRICE_IN_LACS)"]>=60.0)) & (dataset["READY_TO_MOVE"]==1)]
data2
```

### Q.3) Find the owners whose flats under constructions but that flats approved by RERA and show the target price.

```python
data3=dataset.loc[((dataset["POSTED_BY"]=="Owner") & (dataset["UNDER_CONSTRUCTION"]==1) & (dataset["RERA"]==1)) & (dataset["TARGET(PRICE_IN_LACS)"])]
```

### Q.4) Find the owners & dealers those flats size below 900 sqft then increase the target price upto 5 lacs.

```python
data4=dataset.loc[((dataset["POSTED_BY"]=="Owner")| (dataset["POSTED_BY"]=="Dealer")) & (dataset["SQUARE_FT"]<900.000000) & (dataset["TARGET(PRICE_IN_LACS)"]+5.00)]
```

```python
data4
```

```python
AB=(data4["TARGET(PRICE_IN_LACS)"]+5.00) # Create new column of new increasing TARGET(PRICE_IN_LACS).
AB
```

```python
del data4["TARGET(PRICE_IN_LACS)"]           # delete old TARGET(PRICE_IN_LACS) column.
```

```python
data4.insert(11,"TARGET(PRICE_IN_LACS)",AB)  # insert new column of new increasing TARGET(PRICE_IN_LACS). 
```
