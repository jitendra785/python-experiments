```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
Id=np.array([1,2,3,4,5,6,7])
Marks=np.array([65,70,74,80,55,60,75])
plt.bar(Id,Marks)
plt.xlabel("I D")
plt.ylabel("Marks")
plt.title("Marksheet")
plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)
sns.set(style='darkgrid')
plt.show()
```

```python
# To get Inbuilt dataset list of seaborn library
sns.get_dataset_names()
```

```python
# To load dataset from seaborn 
tips= sns.load_dataset('tips')
tips
```

```python
tips.info()
```

```python
sns.relplot(x='total_bill',y='tip', data=tips, hue='day', style='smoker')
```

```python
sns.relplot(x='total_bill',y='tip', data=tips, hue='day', style='smoker',size='time')
```

```python
sns.relplot(x='total_bill',y='tip', data=tips, hue='day', style='smoker',col='time')
```

```python
sns.relplot(x='total_bill',y='tip', data=tips, hue='sex', style='smoker',col='day')
```

```python
sns.relplot(x='total_bill',y='tip', data=tips, hue='sex', style='smoker',col='day', kind='line')
```

```python
sns.catplot(x='smoker',y='tip', data=tips, hue='sex', kind='swarm')
```

```python
sns.lmplot(data=tips, x="total_bill", y="tip", col="time", hue="smoker")
```

```python
sns.displot(data=tips, x="total_bill", col="time", kde=True)
```

```python
sns.catplot(data=tips, kind="violin", x="day", y="total_bill", hue="smoker", split=True)
```
